```python
import polars as pl
from lmf import db
from sage.arith.misc import primes_first_n
from sage.all import libgap
```

```python
def load_check_data(d):
  data = pl.read_csv(f'artin_reps_d{d}_v2.csv', schema_overrides={'Conductor': pl.Int128})
  even_count = data.filter(pl.col('Is_Even') == True).height

  indicator_dict = {}
  grp_dict = {}
  image_dict = {}
  # iterate dataframe rows
  for row in data.iter_rows(named=True):
    indicator = row['Indicator']
    grp = (row['GAP_1'], row['GAP_2'])
    if indicator not in indicator_dict:
      indicator_dict[indicator] = 0
    indicator_dict[indicator] += 1

    if grp not in grp_dict:
      grp_dict[grp] = 0
    grp_dict[grp] += 1

    image = row['Image']
    if image not in image_dict:
      image_dict[image] = 0
    image_dict[image] += 1
  
  print(f'Dimension {d} data check:')
  print(f'  Total records: {data.height}')
  print(f'  Even representations: {even_count}')
  print(f'  Odd representations: {data.height - even_count}')
  print(f'  Indicator counts:')
  for ind, count in indicator_dict.items():
    print(f'    Indicator {ind}: {count}')
  print(f'  GAP group counts:')
  for grp, count in grp_dict.items():
    if grp[0] == 0:
      print(f'    GAP group Unknown: {count}')
    else:
      group = libgap.SmallGroup(grp[0], grp[1])
      print(f'    GAP group {group.StructureDescription()} ({grp[0]}, {grp[1]}): {count}')
  
  print(f'  Image counts:')
  for img, count in image_dict.items():
    print(f'    Image {img}: {count}')

  columns = [f'a_{p:03d}' for p in primes_first_n(168)]
  a_p = data.select(columns)
  # a_p : [rep_counts, 168]
  # a_p가 값이 row나 column 상관 없이 어느 정도 한정된 값만 가지니까, 그 값들하고 그 값이 나오는 횟수를 세보자
  a_p_values = {}
  for row in a_p.iter_rows():
    for val in row:
      if val not in a_p_values:
        a_p_values[val] = 0
      a_p_values[val] += 1
  print(f'  a_p value counts:')
  for val, count in a_p_values.items():
    print(f'    a_p value {val}: {count}')
  
```
